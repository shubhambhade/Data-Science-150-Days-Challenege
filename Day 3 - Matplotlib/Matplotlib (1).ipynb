# 📊 What is Data Visualization?
Data Visualization is the process of representing data graphically using charts, graphs, and maps to make the information easier to understand, interpret, and communicate.

### 📘 What is Matplotlib?

**Matplotlib** is a powerful and versatile **open-source plotting library** for Python, designed to help users visualize data in a variety of formats.

🧑‍💻 **Developed by:** John D. Hunter in **2003**

It enables users to **graphically represent data**, facilitating **easier analysis and understanding**.

🎯 If you want to **convert your boring data into interactive plots and graphs**, Matplotlib is the tool for you.

# What is a Bar Plot?
A bar plot (or bar chart) is used to compare values across categories. The height (or length) of each bar shows the value of that category.

### 📊 `plt.bar()` Parameters – Table Format

| **Parameter**   | **Type**     | **Default** | **Description** |
|------------------|--------------|-------------|------------------|
| `x`              | list or array | —           | Categories or x-coordinates of the bars |
| `height`         | list or array | —           | Height of each bar |
| `width`          | float or list | `0.8`       | Width of the bars |
| `bottom`         | float or list | `None`      | Y-coordinate of bar bases (used for stacked bars) |
| `align`          | str           | `'center'`  | `'center'` or `'edge'` alignment of bars |
| `color`          | str/list      | `None`      | Fill color(s) of the bars |
| `edgecolor`      | str/list      | `None`      | Color of bar borders |
| `linewidth`      | float         | `1`         | Width of bar borders |
| `tick_label`     | list          | `None`      | Custom labels for x-ticks |
| `label`          | str           | `None`      | Label for legend |


```python
#bar plot
x=['python','java','rust','javascript','solidity']
y=[90,70,98,65,43]
plt.xlabel("Popularity of languages", fontsize=15)
plt.ylabel("percentage",fontsize=15)
plt.title("programming languages",fontsize=15)
plt.bar(x,y , width=0.4, color='y', align="center", edgecolor="blue",linewidth=5, linestyle=":", alpha=0.5 , label="popularity") # bydefault center
plt.legend() # apply the label using legend() function
# plt.bar(x,y , width=0.4, color='y', align="edge")
#alpha make the faint colors (0 to 1 range)
plt.show()
```

# What is a Scatter Plot in Matplotlib?
A scatter plot displays individual data points as dots. It is used to visualize relationships or correlations between two numerical variables.

### 🧮 `plt.scatter()`

| **Parameter**  | **Type**        | **Default** | **Description**                                |
|----------------|------------------|-------------|------------------------------------------------|
| `x`            | list/array       | —           | X-axis values                                  |
| `y`            | list/array       | —           | Y-axis values                                  |
| `s`            | scalar/array     | `None`      | Size of each marker (area in points²)          |
| `c`            | color/list       | `None`      | Color of each marker                           |
| `marker`       | str              | `'o'`       | Shape of marker (e.g., `'o'`, `'^'`, `'*'`)     |
| `alpha`        | float [0, 1]     | `None`      | Transparency of markers                        |
| `edgecolors`   | color            | `None`      | Edge color of markers                          |
| `linewidths`   | float            | `None`      | Width of marker edges                          |
| `label`        | str              | `None`      | Label for legend                               |


```python
# scatter plot
x=[1,2,3,4,5]
y=[12,23,32,12,23]
plt.scatter(x,y)
plt.show()
```

```python
# scatter plot
day=[1,2,3,4,5]
no=[12,23,32,12,23]
plt.scatter(day,no)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.show()
```

```python
# scatter plot
day=[1,2,3,4,5]
no=[12,23,32,12,23]
sizes = [200,1400,600,350,670]
plt.scatter(day,no, color=['r','b','y','g','pink'], s= sizes)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.show()
```

# What is a Histogram Plot?
A histogram is a graphical representation of the distribution of numerical data. It groups the data into intervals (bins) and displays the frequency of data points within each interval.

It’s commonly used to understand the shape, spread, and skew of data — especially for continuous variables.

### 📊 `plt.hist()` Parameters

| **Parameter**   | **Type**          | **Default** | **Description**                                        |
|------------------|-------------------|-------------|--------------------------------------------------------|
| `data`           | array/list        | —           | Input numeric data                                     |
| `bins`           | int or sequence   | `10`        | Number or edges of bins                                |
| `color`          | str or list       | `'blue'`    | Fill color of the bars                                 |
| `edgecolor`      | str               | `'black'`   | Border color of bars                                   |
| `alpha`          | float (0 to 1)    | `1.0`       | Transparency level                                     |
| `histtype`       | str               | `'bar'`     | Type of histogram: `'bar'`, `'step'`, `'stepfilled'`   |
| `rwidth`         | float             | `1.0`       | Relative width of bars                                 |
| `label`          | str               | `None`      | Legend label                                           |

#What is a Pie Plot?

A pie plot (or pie chart) is a circular statistical chart divided into slices to illustrate numerical proportions. Each slice represents a category’s percentage share of the total.

### 🥧 `plt.pie()` Parameters
| **Parameter**     | **Type**        | **Default** | **Description**                                         |
|-------------------|------------------|-------------|---------------------------------------------------------|
| `data`            | list/array       | —           | Numerical values (sizes of slices)                      |
| `labels`          | list             | `None`      | Names of each slice                                     |
| `colors`          | list             | `None`      | Colors for slices                                       |
| `autopct`         | str              | `None`      | Format string to display percentage on slices           |
| `startangle`      | float            | `0`         | Rotates the start of the pie (degrees)                  |
| `shadow`          | bool             | `False`     | Add a shadow under the pie                              |
| `explode`         | list             | `None`      | Offsets one or more slices (e.g., `[0, 0.1, 0]`)         |
| `counterclock`    | bool             | `True`      | Direction of slices                                     |


```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["java","python","javascript","solidity"]

plt.pie(x)

plt.show()
```

```python
#labels parameter

x=[10,20,30,40]
y=["java","python","javascript","solidity"]

plt.pie(x, labels=y)

plt.show()
```

```python
#explode parameter
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]

plt.pie(x, labels=y, explode=ex)

plt.show()
```

```python
#colors parameter
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]
colors=["r","g","b","y"]

plt.pie(x, labels=y, explode=ex, colors=colors)

plt.show()
```

```python
#autopct = shows the percentage
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]
colors=["r","g","b","y"]

plt.pie(x, labels=y, explode=ex, colors=colors, autopct="%1.2f%%")

plt.show()
```

```python
# shadow parameter
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]
colors=["r","g","b","y"]

plt.pie(x, labels=y, explode=ex, colors=colors, autopct="%1.2f%%", shadow=True)

plt.show()
```

```python
# radius parameter
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]
colors=["r","g","b","y"]

plt.pie(x, labels=y, explode=ex, colors=colors, autopct="%1.2f%%", shadow=True, radius=1.6, labeldistance=1.1)

plt.show()
```

```python
#rotate  and textprops parameter
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]
colors=["r","g","b","y"]

plt.pie(x, labels=y, explode=ex, colors=colors, autopct="%1.2f%%", shadow=True, labeldistance=1.1, startangle=90, textprops={"fontsize":15})

plt.show()
```

```python
#direction parameter
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]
colors=["r","g","b","y"]

plt.pie(x, labels=y, explode=ex, colors=colors, autopct="%1.2f%%", shadow=True, labeldistance=1.1, startangle=90, textprops={"fontsize":15}, counterclock=False)

plt.show()
```

```python
#wedgeprops and width parameter

#direction parameter
x=[10,20,30,40]
y=["java","python","javascript","solidity"]
ex=[0.3,0.0,0,0]
colors=["r","g","b","y"]

plt.pie(x, labels=y, explode=ex,
        colors=colors, autopct="%1.2f%%", shadow=True, labeldistance=1,
        startangle=90, textprops={"fontsize":15}, counterclock=False ,
        rotatelabels=True)
plt.title("pie chart",fontsize=15)
plt.legend(loc=3)
plt.show()
```

# What is a Stem Plot in Matplotlib?
A stem plot is a vertical line plot that shows data points as stems from a baseline (x-axis) with optional markers on top. It’s useful for visualizing discrete sequences, signal processing, and small datasets.

### 🌱 `plt.stem()` Parameters

| **Parameter**   | **Type**   | **Default** | **Description**                            |
|----------------|------------|-------------|--------------------------------------------|
| `x`            | array/list | —           | X-axis values                              |
| `y`            | array/list | —           | Y-axis values                              |
| `linefmt`      | str        | `'-'`       | Format of the stem lines (color, style)    |
| `markerfmt`    | str        | `'o'`       | Format of the markers on top of stems      |
| `basefmt`      | str        | `'-'`       | Format of the baseline                     |




```python
#stem plot
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]
plt.stem(x,y)
plt.show()
```

### 🎯 `linefmt`

| **Code** | **Type**   | **Examples**             |
|----------|------------|--------------------------|
| `'r'`    | Color      | Red                      |

# What is a Box Plot (Box-and-Whisker Plot)?
A box plot is a standardized way to display the distribution of data based on five summary statistics:

Minimum

First Quartile (Q1)

Median (Q2)

Third Quartile (Q3)

Maximum

It also highlights outliers and skewness in data.

### 📦 `plt.boxplot()` Parameters

| `boxprops`        | dict         | Style of box (e.g., color, linewidth)                     |
| `medianprops`     | dict         | Style of median line                                      |
| `flierprops`      | dict         | Style of outlier points                                   |

# What is a Stack Plot?
A stack plot (or stacked area chart) is used to show how different components contribute to the whole over time. It’s ideal for visualizing trends in data with cumulative totals.

Each category is stacked on top of the previous one, making it easy to compare overall total and individual contributions.

| **Parameter** | **Type**           | **Description**                             |
|---------------|--------------------|---------------------------------------------|
| `x`           | list/array         | Common x-axis values                        |
| `y1, y2, ...` | lists/arrays       | Values for each stacked category            |
| `labels`      | list of str        | Legends for each category                   |
| `colors`      | list of str/colors | Fill colors for each area                   |
| `alpha`       | float (0 to 1)     | Transparency                                |
| `baseline`    | str (optional)     | How the stacks are computed (default is 'zero') |


```python
#2 parameter compulsory pass
#stack plot
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
area=[1,3,2,5,4]
plt.stackplot(x,area)
plt.show()
```

```python
# plot multiple stack plot
x=[1,2,3,4,5]
area1=[1,3,2,5,4]
area2=[1,2,3,4,5]
area3=[5,6,3,2,4]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
#colors parameter
x=[1,2,3,4,5]
area1=[1,3,2,5,4]
area2=[1,2,3,4,5]
area3=[5,6,3,2,4]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l, colors=["r","g","b"],
              baseline="sym") #baseline = zero ,sym ,wiggle
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# plt.grid()
plt.legend()
plt.show()
```

# What is a Step Plot?
A step plot is a line plot that progresses in discrete steps, rather than sloping diagonally between points. It’s useful when you want to:

Visualize changes that occur at specific intervals

Represent piecewise constant functions

Show state transitions (e.g., digital signals, prices after updates)

| **Parameter** | **Type**     | **Description**                                              |
|---------------|--------------|--------------------------------------------------------------|
| `x`           | list/array   | X-axis values                                                |
| `y`           | list/array   | Y-axis values                                                |
| `where`       | str          | Placement of the step: `'pre'`, `'post'`, or `'mid'`         |
| `color`       | str/color    | Line color                                                   |
| `linewidth`   | float        | Thickness of the line                                        |
| `linestyle`   | str          | `'solid'`, `'dashed'`, `'dotted'`, etc.                      |


```python
#step plot
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]
plt.step(x,y, marker='o', color='r', ms=10 ,mfc='g', label="python")
plt.grid()
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc = 4)
plt.show()
```

# What is fill_between() in Matplotlib?
matplotlib.pyplot.fill_between() is used to shade the area between two horizontal curves, often to highlight regions of interest like:

Confidence intervals

Threshold zones

Comparisons between two data lines

| **Parameter** | **Type**     | **Description**                                               |
|---------------|--------------|---------------------------------------------------------------|
| `x`           | array/list   | X values                                                      |
| `y1`          | array/list   | Lower or main boundary                                        |
| `y2`          | array/list   | Upper boundary (default = 0)                                  |
| `where`       | bool array   | Fill only where condition is True                             |
| `interpolate` | bool         | Interpolates when `where` is used                             |
| `step`        | str          | `'pre'`, `'post'`, or `'mid'` for step fill                   |
| `color`       | str          | Fill color                                                    |
| `alpha`       | float        | Transparency of fill (0–1)                                    |
| `label`       | str          | Label for legend                                              |


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5

plt.plot(x, y1, label="Lower bound")
plt.plot(x, y2, label="Upper bound")

# Fill between two curves
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)

plt.title("fill_between plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

# What Is subplot() in Matplotlib?
subplot() lets you draw multiple plots in one figure, organized in a grid of rows and columns.

This is useful to:

Compare multiple visualizations side by side

Analyze different views of the same data

Save space and maintain visual grouping

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

plt.figure(figsize=(10, 8))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("Sin Curve")

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='green')
plt.title("Cos Curve")

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='orange')
plt.title("Tan Curve")

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='red')
plt.title("Exp Decay")

plt.tight_layout()
plt.show()
```

#What is axis in Matplotlib?
In Matplotlib, axis refers to the X and Y coordinate lines and their associated limits, scales, ticks, and labels. It controls what you see and how you see it on a plot.

| Function           | Description                              |
|--------------------|------------------------------------------|
| `plt.axis()`       | Set x/y limits or toggle visibility      |
| `plt.xlim()`       | Set x-axis range                         |
| `plt.ylim()`       | Set y-axis range                         |
| `plt.xlabel()`     | Set label for x-axis                     |
| `plt.ylabel()`     | Set label for y-axis                     |
| `plt.xticks()`     | Set or customize x-axis ticks            |
| `plt.yticks()`     | Set or customize y-axis ticks            |


```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', label="sin(x)")

# Axis labels
plt.xlabel("X Axis →", fontsize=12)
plt.ylabel("Y Axis →", fontsize=12)

# Axis title
plt.title("Axis Control Demo in Matplotlib", fontsize=14)

# Set axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Set custom ticks
plt.xticks([0, 2, 4, 6, 8, 10], labels=["Zero", "Two", "Four", "Six", "Eight", "Ten"])
plt.yticks([-1, 0, 1], labels=["Low", "Mid", "High"])

# Optional: Uncomment to hide or scale axes
# plt.axis('off')        # Hides all axis lines and labels
# plt.axis('equal')      # Equal unit lengths on both axes
# plt.axis('tight')      # Fit tightly to data
# plt.axis('square')     # Makes aspect ratio 1:1

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

#text() in Matplotlib
In Matplotlib, the plt.text() function allows you to add custom text at any (x, y) location in your plot. It's useful for labeling specific points, adding annotations, or writing values directly on the chart.

| Parameter      | Description                          |
|----------------|--------------------------------------|
| `fontsize`     | Size of the text                     |
| `color`        | Text color                           |
| `fontweight`   | `'normal'`, `'bold'`, etc.           |
| `ha` or `horizontalalignment` | `'left'`, `'center'`, `'right'` |
| `va` or `verticalalignment`   | `'top'`, `'center'`, `'bottom'` |
| `rotation`     | Rotate text in degrees               |


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker='o', color='blue')

# Add text at specific data points
plt.text(1, 10, "Start", fontsize=12, color="green")
plt.text(4, 30, "End", fontsize=12, color="red")

# Title and labels
plt.title("Text Annotation Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Plot the data points
plt.plot(x, y, marker='o', color='blue', label="Data Line")

# Loop through each point and add its (x, y) as a label
for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]},{y[i]})", fontsize=9, color="purple", ha='left', va='bottom')

# Add titles and labels
plt.title("Labeled Data Points")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
