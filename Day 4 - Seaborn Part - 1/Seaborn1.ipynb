# Line Plot in Seaborn
Seaborn's lineplot() is a powerful tool to create line charts with statistical confidence intervals, great styling, and support for grouping by categories.

| Parameter   | Description                                               |
|-------------|-----------------------------------------------------------|
| `data`      | DataFrame or data source                                  |
| `x`, `y`    | Column names for X and Y axes                             |
| `hue`       | Grouping variable that changes line color                 |
| `style`     | Different line styles for categories                      |
| `markers`   | Use markers on data points                                |
| `ci`        | Confidence interval (default = 95%, use `ci=None` to turn off) |
| `palette`   | Set color scheme (`"muted"`, `"bright"`, etc.)            |


Syntax: lineplot(x,y,data)


where,


x- data variable for x-axis


y- data variable for y-axis


data- data to be plotted

```python
import pandas as pd

var1 = [1,2,3,4,5,6,7]
var2 = [5,6,7,1,2,3,4]
df = pd.DataFrame({'x':var1,'y':var2})
sns.lineplot(x="x",y="y" , data=df)
plt.show()
```

# Bar Plot


```python
sns.barplot(x="island",y="bill_length_mm",data=dataframe)
plt.show()
```

```python
#hue parameter
sns.barplot(x="island",y="bill_length_mm",data=dataframe , hue="sex")
plt.show()
```

```python
#order parameter
sns.barplot(x="island",y="bill_length_mm",data=dataframe , hue="sex",
            order=["Torgersen","Dream","Biscoe"], hue_order=['Female','Male'])
plt.show()
```

```python
# ci parameter- shows interval (range of value 0 to 100) -> deprecated
# new option = errorbar=('ci', 100)
sns.barplot(x="island",y="bill_length_mm",data=dataframe , hue="sex",
            order=["Torgersen","Dream","Biscoe"], hue_order=['Female','Male'],
            errorbar=('ci', 100), n_boot=3)
plt.show()
```

```python
#orient parameter - make the graph horizontal or vertical. but when we make graph horizontal x and y parameter both are numerical. otherwise get error.
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=dataframe ,
            orient="h", color='r')
plt.show()
```

```python
#palette parameter
sns.barplot(x="island",y="bill_length_mm",data=dataframe ,
             palette="icefire",hue="sex")
plt.show()
```

```python
# saturation parameter -> it reduces the hardness of color make blur.
# errcolor deprecated -> change the color of interval line
sns.barplot(x="island",y="bill_length_mm",data=dataframe ,
             palette="icefire",hue="sex",saturation=0.3 ,err_kws={'color': 'b'})
plt.show()
```

```python
# capsize
sns.barplot(x="island",y="bill_length_mm",data=dataframe ,
             hue="sex" , capsize=0.5)
plt.show()
```

```python
#dodge parameter -  seperate the overlapping bars
sns.barplot(x="island",y="bill_length_mm",data=dataframe ,
             hue="sex" , dodge=False)
plt.show()
```

```python
#alpha parameter - make the graphs color light.
#seaborn grid
sns.set(style="darkgrid")

sns.barplot(x="island",y="bill_length_mm",data=dataframe ,
             hue="sex" , dodge=True, alpha=0.5)
plt.show()
```

# Histograms

Histograms are visualization tools that represent the distribution of a set of continuous data. In a histogram, the data is divided into a set of intervals or bins (usually on the x-axis) and the count of data points that fall into each bin corresponding to the height of the bar above that bin. These bins may or may not be equal in width but are adjacent (with no gaps).  

A density plot (also known as kernel density plot) is another visualization tool for evaluating data distributions. It can be considered as a smoothed histogram. The peaks of a density plot help display where values are concentrated over the interval. There are a variety of smoothing techniques. Kernel Density Estimation (KDE) is one of the techniques used to smooth a histogram.

```python
import seaborn as sn
import matplotlib.pyplot as plt
df = sn.load_dataset("penguins")
sn.displot(df["flipper_length_mm"])
plt.show()
```

```python
#bins parameter
sn.displot(dataframe["flipper_length_mm"]
           ,bins=[170, 180,190,200,210,220,230,240])
plt.show()
```

```python
#kde - kernal density estimater -> above the bar graph show the line graph.
sn.displot(dataframe["flipper_length_mm"]
           ,bins=[170, 180,190,200,210,220,230,240],kde=True)
plt.show()
```

```python
#rug -> Adds vertical ticks at each data point (along x-axis)
sn.displot(dataframe["flipper_length_mm"]
           ,bins=[170, 180,190,200,210,220,230,240],kde=True,rug=True , color='g')
```

```python
# log_scale
sn.displot(dataframe["body_mass_g"]
           ,log_scale=True)
```

# Seaborn heatmap() â€” Visualizing Matrix-Like Data

A heatmap is a powerful visualization tool for displaying matrix-like data, where color intensity represents data values.

A heatmap is a graphical representation of data where individual values are represented by color intensity. It is widely used in data analysis and visualization to identify patterns, correlations and trends within a dataset. Heatmaps in Seaborn can be plotted using the seaborn.heatmap() function,

| **Parameter**   | **Description**                                                                 |
|-----------------|----------------------------------------------------------------------------------|
| `data`          | 2D dataset (NumPy array, DataFrame, or matrix)                                   |
| `annot`         | If `True`, write the data values in each cell                                   |
| `fmt`           | String formatting for annotations (`.2f`, `.0f`, `.1g`, etc.)                    |
| `cmap`          | Colormap used for shading (`'coolwarm'`, `'YlGnBu'`, `'magma'`, etc.)           |
| `linewidths`    | Width of the lines that separate cells                                           |
| `linecolor`     | Color of the lines that separate cells                                           |
| `vmin`, `vmax`  | Minimum and maximum values for the color scale                                   |
| `cbar`          | Show color bar (default is `True`)                                               |
| `square`        | If `True`, set cells to be square                                                |
| `xticklabels`   | Control display of column labels (`True`, `False`, or list of labels)            |
| `yticklabels`   | Control display of row labels (`True`, `False`, or list of labels)               |
| `mask`          | Boolean array same shape as `data`; cells with `True` are hidden                 |
| `center`        | Value at which to center the colormap                                            |


```python
import numpy as np
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
sn.heatmap(var, vmin=0, vmax = 10, annot=True)
plt.show()
#annot -> displays the values in heatmap boxes.
```
