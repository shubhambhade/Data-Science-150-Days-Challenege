# Count Plot :

seaborn.countplot() is a function in the Seaborn library in Python used to display the counts of observations in categorical data. It shows the distribution of a single categorical variable or the relationship between two categorical variables by creating a bar plot.

###  Count Plot Parameters

| Parameter    | Meaning                                                |
|--------------|--------------------------------------------------------|
| `x`, `y`     | Columns to plot on X and Y axes                        |
| `hue`        | Adds color based on a third column (for grouping)      |
| `data`       | The dataset (usually a DataFrame)                      |
| `order`      | Order of categories on the axis                        |
| `hue_order`  | Order of hue categories (colors)                       |
| `orient`     | Plot direction: `"v"` = vertical, `"h"` = horizontal   |
| `color`      | Set a single color for all bars/plots                 |
| `palette`    | Set different colors for each category/group           |
| `saturation` | Intensity of color (0 to 1)                            |
| `dodge`      | Shift elements when using `hue` (True/False)           |
| `ax`         | Use a specific Matplotlib Axes to draw the plot        |
| `**kwargs`   | Extra settings for more control                        |


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataframe = sns.load_dataset('tips')
dataframe
```

```python
sns.countplot(x='day',data=dataframe,  hue="day" )
plt.show()
```

```python
sns.countplot(x='sex',data=dataframe, hue="smoker",
              palette="Accent", saturation=0.5)
plt.show()
```

# Violinplot

A violin plot is a statistical graph that combines the features of a box plot and a kernel density plot. It displays the distribution of quantitative data across one or more categorical variables. Violin plots not only show summary statistics (like median and interquartile range) but also the probability density of the data at different values, providing more insight into the data distribution.

They are particularly useful for comparing multiple groups and are more informative than box plots when the shape of the distribution is important.

```python
sns.violinplot(x='day', y='total_bill', data=dataframe , hue="day")
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=dataframe , hue="time", linewidth=2)
plt.show()
```

ðŸ”¸color parameter :  

The color parameter is used to set the same color for all elements in the plot,
 ignoring the hue grouping if present.

```python
sns.violinplot(x='day', y='total_bill', data=dataframe , color="red")
plt.show()
```

ðŸ”¸ palette Parameter

Use: Sets different colors for categories (used with hue).

âœ… Examples of palette values:
'deep' â†’ Default colorful palette

'pastel' â†’ Soft, light colors

'bright' â†’ Vivid and strong colors

'muted' â†’ Softer than bright, more serious tones

'dark' â†’ Darker tones

'colorblind' â†’ Friendly for colorblind users

["red", "blue"] â†’ Custom list of colors

{"Male": "blue", "Female": "pink"} â†’ Custom dictionary mapping for specific hue values

```python
#palette parameter
sns.violinplot(x='day', y='total_bill', data=dataframe , palette="Dark2")
plt.show()
```

ðŸ”¸ order parameter =  assign the order of elements by passing list

ðŸ”¸ saturation = make the graphs color light or dark (range- 0 for lighter to 1 for dark.

```python

sns.violinplot(x='day', y='total_bill', data=dataframe
               , order=["Sat","Sun","Thur","Fri"],hue="day", saturation=0.5)
plt.show()
```

ðŸ”¸ split parameter

Use: Splits the violin into two halves for each hue category.
Only works when hue is set.

âœ… Helps compare two groups (like Male vs Female) side by side in the same violin.

ðŸ”¸ scale parameter

Use: Controls the width/size of each violin.

Options:

'area' (default): All violins have the same area.

'count': Width depends on the number of observations.

'width': All violins have the same width.

âœ… Makes violins wider or narrower based on how many data points are in each group.

```python
sns.violinplot(x='day', y='total_bill', data=dataframe ,
               hue="sex", split=True,
               scale='count')
plt.show()
```

```python
#plot horizontal graph change the x and y axis values
sns.violinplot(x='total_bill', y='day', data=dataframe , hue="sex",
               split=True,
               scale='count')
plt.show()
```

ðŸ”¸inner parameter:

This will show quartile lines inside each violin instead of a box.

inner values :

1] 'box' â†’ Draws a mini boxplot inside the violin.

2] 'quartiles' â†’ Displays the 25%, 50%, and 75% quartile lines.

3] 'point' â†’ Shows each individual data point as a dot.

4] 'stick' â†’ Shows individual data points as vertical lines (sticks).

5] None â†’ Draws nothing inside the violin.

```python
sns.violinplot(x='time', y='total_bill', data= dataframe, inner='box',
               palette="Dark2")
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data= dataframe, inner='quartiles',
               palette="Dark2")
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data= dataframe, inner='point',
               palette="Dark2")
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data= dataframe, inner='stick',
               palette="Dark2")
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data= dataframe, inner=None,
               palette="Dark2")
plt.show()
```

```python
# create single violinplot along x-axis
sns.violinplot(x=dataframe['time'])
plt.show()
```

```python
# create single violinplot along y-axis
sns.violinplot(y=dataframe['time'])
plt.show()
```

#pair plot

The seaborn.pairplot() method is used to visualize relationships between multiple variables in a dataset.
It creates a grid of scatter plots and histograms to help identify patterns, correlations, and trends.

```python
sns.pairplot(dataframe)
plt.show()
```

```python
sns.pairplot(dataframe, hue='sex')
plt.show()
```

```python
#vars = select only required data column to plotting.
#hue_order = set the order which one come first , second that type
# palette parameter
sns.pairplot(dataframe , vars=['total_bill','tip'],
             hue='sex', hue_order=["Female","Male"],
             )
plt.show()
```

```python
#x_vars =  set x-axis variables.
#y_vars = set y-axis variables.
sns.pairplot(dataframe , x_vars=['total_bill','tip'],
             hue='sex', hue_order=["Female","Male"],
             palette="Dark2"
             )
plt.show()
```

```python
#x_vars =  set x-axis variables.
#y_vars = set y-axis variables.
sns.pairplot(dataframe , y_vars=['total_bill','tip'],
             hue='sex', hue_order=["Female","Male"],
             palette="Dark2"
             )
plt.show()
```

```python
#kind parameter = define type of graph. (reg , kde , hist, scatter)
#diag_kind = define diagonal graph types. (reg , kde , hist, scatter)
sns.pairplot(dataframe ,
             hue='sex', hue_order=["Female","Male"],
             palette="Dark2",
             kind="reg",
             diag_kind="hist"
             )
plt.show()
```

```python
#markers = give symbols to present different categories.
# plot_kws= set the size of markers,
data = dataframe[:10]
sns.pairplot(data, hue='sex',
             markers=["*", "<"],     # Different markers for Male and Female
             plot_kws={'s': 120 })     # Set marker size for all points
plt.show()
```

# Strip Plot

A strip plot is a scatter plot where categorical data is plotted along one axis and numeric data along the other.

It shows all individual data points â€” useful when you want to see data distribution and overlapping points.


### ðŸ“‹ `seaborn.stripplot()` â€“ Parameters Table

| **Parameter** | **Use**                                                                 |
|----------------|------------------------------------------------------------------------|
| `x`, `y`        | Categorical and numerical variables to plot.                          |
| `data`          | Dataset to use (usually a pandas DataFrame).                          |
| `hue`           | Group data points by a category using different colors.               |
| `jitter`        | Spread points horizontally to avoid overlap (default is `True`).      |
| `dodge`         | Separate hue groups side by side (works with `jitter=True`).          |
| `palette`       | Color scheme for different hue levels.                                |
| `size`          | Controls the size of the markers (dots).                              |
| `marker`        | Sets the shape of the data points (e.g., `"o"`, `"s"`, `"^"`).         |




```python
sns.stripplot(data=dataframe)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill',
              data=dataframe,hue='sex')
plt.show()
```

```python
#linewidth - set the width of data points.
#edgecolor - set the color of edge  of data points.
data1=dataframe[:100]
sns.stripplot(x='day', y='total_bill',
              data=data1,hue='sex',
              linewidth=2,
              edgecolor='black')
plt.show()
```

```python
#jitter parameter
sns.stripplot(x='day', y='total_bill',
              data=dataframe,hue='sex',
              jitter=10)
plt.show()
```

```python
# size parameter - set the size of data points
sns.stripplot(x='day', y='total_bill',
              data=dataframe,hue='sex',
              size=10)
plt.show()
```

```python
#marker and alpha parameter
sns.stripplot(x='day', y='total_bill',
              data=data1,hue='sex',
              marker="*",
              alpha=0.8)
plt.show()
```

```python
sns.stripplot(x= dataframe['total_bill'])
plt.show()
```

```python
sns.stripplot(y= dataframe['total_bill'])
plt.show()
```

#ðŸ“¦ Box Plot & Whisker Plot in Seaborn

In Seaborn, the box plot (also called a box-and-whisker plot) is used to show the distribution, median, and spread of a dataset. It highlights key statistics like minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum, along with outliers.



```python
sns.set(style="whitegrid")
sns.boxplot(x='day', y='total_bill', data=dataframe,
            )
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day', y='total_bill', data=dataframe,
           hue='time')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day', y='total_bill', data=dataframe,
           hue='sex',
            order=['Sat','Thur','Sun','Fri'])
plt.show()
```

```python
# showmeans parameter
#orient =(v , h) -> we make horizontal graph that time give data to
#both axis numerical only otherwise give error.
sns.set(style="whitegrid")
sns.boxplot(x='day', y='total_bill',
            data=dataframe,
           hue='sex', showmeans=True,
            color='r',
            meanprops={'marker':"+",'markeredgecolor' :'b'},
            linewidth=3,
            palette="Accent",
            orient='v')
plt.show()
```

```python
# showmeans parameter
#orient =(v , h) -> we make horizontal graph that time give data to
#both axis numerical only otherwise give error.
sns.set(style="whitegrid")
sns.boxplot( data=dataframe,
          showmeans=True,
            color='r',
            meanprops={'marker':"+",'markeredgecolor' :'b'},
            linewidth=3,
            palette="Accent",
            orient='h')
plt.show()
```

```python
#create single box plot
sns.boxplot(data=dataframe['total_bill'])
plt.show()
```

ðŸ“Š factorplot() in Seaborn (Now Deprecated)

seaborn.factorplot() was a powerful function used to draw categorical plots like bar plots, box plots, strip plots, etc., with support for multiple facets (subplots) using col, row, and hue.

However, it's now deprecated and replaced by ðŸ‘‰ **seaborn.catplot()** in recent versions of Seaborn.

### ðŸ“‹ `seaborn.catplot()` â€“ Parameters Table

| **Parameter** | **Use** |
|---------------|--------|
| `data`        | Dataset to plot (usually a DataFrame). |
| `x`, `y`      | Categorical and numerical variables. |
| `hue`         | Grouping variable for color encoding. |
| `col`         | Creates separate subplots for each unique value in this column. |
| `row`         | Creates subplots across rows. |
| `kind`        | Type of plot: `'strip'`, `'swarm'`, `'box'`, `'violin'`, `'bar'`, `'point'`. |
| `palette`     | Colors used for different hue levels. |
| `order`       | Custom order of categories on the axis. |
| `hue_order`   | Custom order of hue categories. |
| `col_order`   | Order to display subplot columns. |
| `row_order`   | Order to display subplot rows. |
| `height`      | Height of each facet/subplot (in inches). |
| `aspect`      | Aspect ratio of each facet (width = height Ã— aspect). |
| `margin_titles` | Set subplot titles on the margins (True/False). |
| `legend`      | Show or hide legend (`True`/`False`). |
| `legend_out`  | Place legend outside the plot (`True`/`False`). |
| `sharex`, `sharey` | Share x or y axis scale across plots (`True`/`False`). |
| `facet_kws`   | Extra options passed to the `FacetGrid` object. |


```python
#create cat plot
sns.catplot(data=dataframe)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=dataframe,
            hue='time')
plt.show()
```

# Seaborn Figure Styles

This affects things like the color of the axes, whether a grid is enabled by default, and other aesthetic elements.

The ways of styling themes are as follows:

white
dark
whitegrid
darkgrid
ticks

```python
# create box plot
sns.set_theme(style="whitegrid")
sns.boxplot(x='day', y='total_bill', data=dataframe)
plt.show()
```

```python
#another way
sns.set_style("darkgrid")
sns.barplot(x='day', y='total_bill', data=dataframe
            ,hue='sex')
plt.show()
```

```python
#another way
sns.set_style("dark")
sns.barplot(x='day', y='total_bill', data=dataframe
            ,hue='sex')
plt.show()
```

```python
#another way
sns.set_style("ticks")
sns.violinplot(x='day', y='total_bill', data=dataframe
            ,hue='sex')
plt.show()
```

```python
#remove axis - despine()
sns.set_style("whitegrid")
plt.figure(figsize=(20,8)) # set the size of figure
sns.barplot(x='day', y='total_bill', data=dataframe
            ,hue='sex')
sns.despine()
plt.show()
```

```python
#change content
sns.set_style("whitegrid")
sns.set_context("talk")
sns.barplot(x='day', y='total_bill', data=dataframe
            ,hue='sex',
            palette='cool')
plt.show()
```
